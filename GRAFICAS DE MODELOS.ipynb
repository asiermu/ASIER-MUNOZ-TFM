## GOMPERTZ 

```python
import numpy as np
import matplotlib.pyplot as plt
import random


# Definir el rango de tiempo
tiempo = np.linspace(0, 30, 1000)  # De 0 a 10 unidades de tiempo
r=0.3
N0=10
K=1000
C = np.log(-np.log(N0 / K)) + r * 0  # Suponiendo t0=0
    
    # Calcular N en función del tiempo utilizando la ecuación de Gompertz
N = K * np.exp(-np.exp(-r * tiempo + C))
    # Graficar la curva de Gompertz
plt.plot(tiempo, N, label=f' Gompertz : r={r:.2f}, N0={N0}, K={K}')

# Añadir leyenda
plt.legend()
plt.title('Gompertz')
plt.xlabel('Tiempo')
plt.ylabel('Tamaño')
plt.savefig('Función Gompertz.png')
plt.show()
```
